# GAUSSIAN BLUR

```python
img2=cv.imread("C:/Users/admin/Pictures/cv/hgfv.jfif")
```

```python
gblur=cv.GaussianBlur(img2,(5,5),0)
cv.imshow("image",img2)
cv.imshow("blur",gblur)
cv.waitKey(0)
cv.destroyAllWindows()
```

# BILATERAL FILTER

```python
bilblur=cv.bilateralFilter(img2,5,150,150)
cv.imshow("image",img2)
cv.imshow("blur",bilblur)
cv.waitKey(0)
cv.destroyAllWindows()
```

# MEDIAN FILTER

# LAPLACIAN  FILTER

```python
butterfly=cv.imread("C:/Users/admin/Pictures/cv/buterrfly.jpg",0)
butterfly=cv.cvtColor(butterfly,cv.COLOR_BGR2RGB)
```

```python
lap=cv.Laplacian(butterfly,cv.CV_64F,ksize=9)
lap
cv.imshow("LAP",lap)
cv.imshow("IMG",butterfly)
cv.waitKey(0)
cv.destroyAllWindows()
```

# SOBEL FILTER

```python
sobx=cv.Sobel(butterfly,cv.CV_64F,1,0,ksize=5)
soby=cv.Sobel(butterfly,cv.CV_64F,0,1,ksize=5)
sobx1=np.uint8(np.absolute(sobx))
cv.imshow("image",butterfly)
cv.imshow("sobel 0",sobx)
cv.imshow("sobel 1",soby)
cv.imshow("sobel 2",sobx1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
titles=["image", "laplacian","sobelx","sobely"]
img=[butterfly,lap,sobx,soby]
for i in range(len(img)):
    plt.subplot(2,2,i+1),plt.imshow(img[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
